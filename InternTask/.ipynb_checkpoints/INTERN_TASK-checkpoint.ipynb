```python
import os
import random
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
path='X-ray_image/'
ids=random.sample(next(os.walk(path))[2],6)
```

```python
ids
```

```python
from tqdm import tqdm
from skimage.io import imshow,imread
import cv2
from skimage.transform import resize

imgs=[]

for n,id_ in tqdm(enumerate(ids),total=len(ids)):
    tpath=path+id_
    imgs.append(cv2.imread(tpath)[:,:,:3])
    
    
```

```python
type(imgs)
```

```python
imgs=[resize(img,(256,256),mode='constant',preserve_range=True) for img in imgs]
print(imgs[0].shape)
```

```python
imgs
```

```python
imgs=np.asarray(imgs)
imgs
```

```python
type(imgs)
```

```python
imgs.shape
```

```python
#rescale and resize

from skimage.transform import rescale,resize,downscale_local_mean
rescaled_imgs=[]
for img in imgs:
    rescaled_imgs.append(rescale(img,1.0/4.0,anti_aliasing=True))

```

```python
rescaled_imgs=np.asarray(rescaled_imgs)
rescaled_imgs
```
