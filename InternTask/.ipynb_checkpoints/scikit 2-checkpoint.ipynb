```python
import numpy as np
import os
import  random
import matplotlib.pyplot as plt
from skimage import io,restoration
from skimage.filters.rank import entropy
from skimage.morphology import disk
```

```python
path='X-ray_image/'
ids=next(os.walk(path))[2]
num=random.randint(0,len(ids))
img=io.imread(path+ids[num],as_gray=True)
entropy_img=entropy(img,disk(3))
plt.imshow(entropy_img,cmap='gray')
```

```python
from skimage.filters import try_all_threshold
```

```python

fig,axes=try_all_threshold(entropy_img,figsize=(10,8),verbose=False)
plt.show()
```

```python
#Otsu

from skimage.filters import threshold_otsu
thresh=threshold_otsu(entropy_img)
thresh
```

```python
binary_otsu=entropy_img<=thresh
```

```python
from skimage.filters import threshold_yen
thresh=threshold_yen(entropy_img)
thresh
```

```python
binary_yen=entropy_img<=thresh
```

```python
plt.imshow(binary_otsu,cmap="gray")
```

```python
plt.imshow(binary_yen,cmap="gray")
```

```python
print("The percentage bright pixels of otsu is : ",(np.sum(binary_otsu==1)*100)/(np.sum(binary_otsu==1)+np.sum(binary_otsu==0)))
```

```python
print("The percentage bright pixels of otsu is : ",(np.sum(binary_yen==1)*100)/(np.sum(binary_yen==1)+np.sum(binary_yen==0)))
```
