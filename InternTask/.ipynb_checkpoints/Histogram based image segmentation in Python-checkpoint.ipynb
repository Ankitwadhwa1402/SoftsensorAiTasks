```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from skimage import io,img_as_float,img_as_ubyte
```

```python
from skimage.restoration import denoise_nl_means,estimate_sigma
```

```python
img=img_as_float(io.imread('test.jpg'))
plt.imshow(img,cmap="gray")
```

```python
sigma_est=np.mean(estimate_sigma(img,channel_axis=None))
```

```python
denoise=denoise_nl_means(img,h=1.15*sigma_est,fast_mode=True,patch_size=5,patch_distance=3,channel_axis=None)
```

```python
denoise_ubyte=img_as_ubyte(denoise)
plt.imshow(denoise,cmap="gray")
```

```python
plt.hist(denoise_ubyte.flat,bins=100,range=(0,255))
plt.show()
```

```python
segm1=(denoise_ubyte<=80)
segm2=(denoise_ubyte>80) & (denoise_ubyte<=120)
segm3=(denoise_ubyte>120) & (denoise_ubyte<=250)
```

```python
all_segments=np.zeros((denoise_ubyte.shape[0],denoise_ubyte.shape[1],3))
```

```python
all_segments[segm1]=(1,0,0)
all_segments[segm2]=(0,1,0)
all_segments[segm3]=(0,0,1)

plt.imshow(all_segments)
```

```python
from scipy import ndimage as nd
segm1_opened=nd.binary_opening(segm1,np.ones((5,5)))
segm1_closed=nd.binary_closing(segm1_opened,np.ones((5,5)))

segm2_opened=nd.binary_opening(segm2,np.ones((5,5)))
segm2_closed=nd.binary_closing(segm2_opened,np.ones((5,5)))

segm3_opened=nd.binary_opening(segm3,np.ones((5,5)))
segm3_closed=nd.binary_closing(segm3_opened,np.ones((5,5)))
```

```python
all_segments_cleaned=np.zeros((denoise_ubyte.shape[0],denoise_ubyte.shape[1],3))
all_segments_cleaned[segm1_closed]=(1,0,0)
all_segments_cleaned[segm2_closed]=(0,1,0)
all_segments_cleaned[segm3_closed]=(0,0,1)
```

```python
plt.imshow(all_segments_cleaned)
plt.imsave("segmented_img.jpg",all_segments_cleaned)
```
