# Deconvolution



```python
import numpy as np
    
psf=np.ones((2,2))/4  # / by 9 be matrix is of 9 1's and we don't want to increase the energy of img we want to * by 1 :. / 9; 
```

```python
import scipy.stats as st
def gkern(kernlen=21, nsig=2):
    lim=kernlen//2+(kernlen%2)/2
    x = np.linspace(-lim, lim, kernlen+1)
    kern1d = np.diff(st.norm.cdf(x))
    kern2d = np.outer(kern1d, kern1d)
    return kern2d/kern2d.sum()
psf_gkern=gkern(5,3)
print(psf_gkern)
```

```python
#sharpening the img
from skimage import restoration
deconv_img,_=restoration.unsupervised_wiener(img,psf_gkern) #psf is the value if we know why our image is blur in mathematical way 
# _ point spread function
```
