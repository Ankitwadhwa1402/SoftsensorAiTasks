```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io,img_as_float,img_as_ubyte
```

```python
from skimage.restoration import denoise_nl_means,estimate_sigma
```

```python
img=img_as_float(io.imread('test.jpg'))
plt.imshow(img,cmap="gray")
```

```python
sigma_est=np.mean(estimate_sigma(img,channel_axis=None))
```

```python
denoise_img=denoise_nl_means(img,h=1.15*sigma_est,fast_mode=True,patch_size=5,patch_distance=3,channel_axis=None)
```

```python
denoise_img=img_as_ubyte(denoise)
plt.imshow(denoise_img,cmap="gray")
```

```python
plt.hist(denoise_img.flat,bins=100,range=(0,1))
plt.show()
```

```python
from skimage import exposure
```

```python
eq_img=exposure.equalize_adapthist(denoise_img)
```

```python
plt.hist(eq_img.flat,bins=100,range=(0,255))
plt.show()
```
