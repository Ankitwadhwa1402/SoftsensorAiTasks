```python
import cv2
import numpy as np
from skimage import io,img_as_float
from skimage.filters import gaussian
import matplotlib.pyplot as plt
```

```python
img_gaussian_noise=img_as_float(io.imread('test.jpg',as_gray=True))
img_salt_pepper=img_as_float(io.imread('test.jpg',as_gray=True))

img=img_gaussian_noise
```

```python
gaussian_kernel=np.array([[1/16,1/8,1/16],
                         [1/8,1/4,1/8],
                         [1/16,1/8,1/16]])
conv_using_cv2=cv2.filter2D(img,-1,gaussian_kernel,borderType=cv2.BORDER_CONSTANT)
```

```python
plt.imshow(img,label="original")
# cv2.imshow(conv_using_cv2,label="cv2 filters",)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
plt.show()
```

```python
plt.imshow(conv_using_cv2,label="cv2 filters",cmap="gray")
```

```python
gaussian_using_cv2=cv2.GaussianBlur(img,(3,3),0,borderType=cv2.BORDER_CONSTANT)
plt.imshow(gaussian_using_cv2,label="gaussian filters",cmap="gray")
```

```python
gaussian_using_skimage=gaussian(img,sigma=3,mode='constant',cval=0.0)
plt.imshow(gaussian_using_skimage,label="sk filters",cmap="gray")
```
