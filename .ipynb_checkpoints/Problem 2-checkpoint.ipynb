# Average loan for a particular State.

```python
tempDf=df[['CurrentApprovalAmount','BorrowerState']]
tempDf.info()
```

```python
tempDf=tempDf.dropna()
```

```python
tempDf.groupby('BorrowerState').mean()
```

```python
tempDf = tempDf.astype({"BorrowerState": str, "CurrentApprovalAmount": int}) 
```

```python
plt.figure(figsize=(18,4), dpi=80)
plt.bar(tempDf['BorrowerState'],tempDf['CurrentApprovalAmount'],color='cyan')
plt.title('Bar PLOT')
plt.xlabel('BorrowerState')
plt.ylabel('CurrentApprovalAmount')
plt.show()
```

# Number of jobs reported by different business types.

```python
tempDf=df[['JobsReported','BusinessType']]
tempDf=tempDf.groupby('BusinessType').sum()
tempDf
```

```python
tempDf.nunique()
```

```python
tempDf.shape
```

```python
tempDf=tempDf.dropna()
```

```python
tempDf.shape
```

```python
plt.figure(figsize=(8,8), dpi=80)
plt.pie(tempDf['JobsReported'],startangle=90)
plt.legend( labels=tempDf.index, bbox_to_anchor=(-0.2, 1.),
           fontsize=16)
plt.show()
```

# Highest loan lender in each city.

```python
tempDf=df[['BorrowerState','ServicingLenderName','CurrentApprovalAmount']].groupby('BorrowerState').max()
```

```python
tempDf.sort_values('CurrentApprovalAmount',ascending=False,inplace=True)
```

```python
import seaborn as sn
```

```python
sns.set(rc={"figure.figsize":(17, 7)}) 
sns.barplot(x='ServicingLenderName',y='CurrentApprovalAmount',data=tempDf.head(5))
```
